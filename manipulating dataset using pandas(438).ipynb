```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv("India_rainfall.csv")
data
```

```python
print(data.describe())
```

```python
a=data.head(10)
print(a)
```

```python
filtered_df = a[a['Actual Rainfall: JUN']>105]
print("\nFiltered Data:")
print(filtered_df)
```

```python
filtered_data=a[a['Actual Rainfall: JUN']<200]
print(filtered_data)
```

```python
sorted_data = a.sort_values('Actual Rainfall: JUN')
print("Sorted Data:")
print(sorted_data)
```

```python
sorted_data=a.sort_values('Actual Rainfall: JUN',ascending=True)
print(sorted_data)
```

```python
#sorted_data=a.sort_values('Actual Rainfall: JUN',descending=True)
```

```python
sorted_data=a.sort_values('Actual Rainfall: JUN',ascending=False)
print(sorted_data)
```

```python
grouped_data = sorted_data.groupby('YEAR')#.mean()
print("Grouped Data:")
print(grouped_data)
```

```python
grouped_data = sorted_data.groupby('YEAR').mean()
print("Grouped Data:")
print(grouped_data)
```
